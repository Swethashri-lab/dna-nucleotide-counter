```python
# ============================================
# DAY 4 — COMPLETE FASTA ANALYSIS
# Everything in one cell — no dependency issues
# ============================================


# ---- STEP 1: CREATE THE FASTA FILE ----

fasta_content = """>seq1 Human BRCA1 fragment
ATGGATTTATCTGCTCTTCGCGTTGAAGAAGTACAAAATGTCATTAATGCTATGCAGA
AAATCTTAGAGTGTCCCATCTGTCTGGAGTTGATCAAGGAACCTGTCTCCACAAAGTGT
>seq2 Human TP53 tumor suppressor
ATGGAGGAGCCGCAGTCAGATCCTAGCGTTGAATCTGAGGAATGCTTCGTCCTTTCACT
CCATCGATGCTGTTCCGGGAGCAGATAAGCGGGCCGGTCTTCCCTTCCCAGAAACCCCT
>seq3 Mouse beta-actin
ATGGATGATGATATCGCCGCGCTCGTCGTCGACAACGGCTCCGGCATGTGCAAAGCCGG
CTTCGCGGGCGACGATGCCCCGAGGGCCGTCTTCCCCTCCATCGTGGGGCGCCCCAGGC"""

with open("sample.fasta", "w") as file:
    file.write(fasta_content)

print("✅ STEP 1: sample.fasta created!")


# ---- STEP 2: VERIFY FILE EXISTS ----

print()
print("✅ STEP 2: Raw file contents:")
print("-" * 40)

with open("sample.fasta", "r") as file:
    content = file.read()
    print(content)

print("-" * 40)


# ---- STEP 3: PARSE THE FASTA FILE ----

def parse_fasta(filename):

    sequences    = {}
    current_name = ""
    current_seq  = ""

    with open(filename, "r") as file:

        for line in file:

            line = line.strip()

            if line == "":
                continue        # skip empty lines completely

            if line.startswith(">"):

                if current_name != "":
                    sequences[current_name] = current_seq

                current_name = line[1:]
                current_seq  = ""

            else:
                current_seq = current_seq + line

        if current_name != "":
            sequences[current_name] = current_seq

    return sequences


print()
print("✅ STEP 3: Parsing complete!")
sequences = parse_fasta("sample.fasta")
print("Total sequences found:", len(sequences))


# ---- STEP 4: FULL ANALYSIS OF EVERY SEQUENCE ----

def analyze_fasta(sequences):   # takes dictionary directly, not filename

    complement_map = {
        "A" : "T",
        "T" : "A",
        "G" : "C",
        "C" : "G"
    }

    print()
    print("=" * 55)
    print("     COMPLETE FASTA ANALYSIS REPORT")
    print("=" * 55)
    print("Total sequences:", len(sequences))
    print("=" * 55)

    for name, dna in sequences.items():

        dna = dna.upper().strip()

        # Skip if sequence is empty
        if len(dna) == 0:
            print("⚠️ Skipping empty sequence:", name)
            continue

        # Day 1 — nucleotide counts
        a = dna.count("A")
        t = dna.count("T")
        g = dna.count("G")
        c = dna.count("C")

        # Day 2 — GC content
        gc = round((g + c) / len(dna) * 100, 2)

        # Day 3 — transcription
        rna = dna.replace("T", "U")

        # Day 3 — reverse complement
        complement = ""
        for base in dna:
            if base in complement_map:          # safety check
                complement = complement + complement_map[base]
        rev_comp = complement[::-1]

        # Print report
        print()
        print("🧬 SEQUENCE  :", name)
        print("   Length    :", len(dna), "bases")
        print("   A:", a, "  T:", t, "  G:", g, "  C:", c)
        print("   GC Content:", gc, "%")

        if gc > 60:
            print("   Stability  : 🔴 High GC — very stable")
        elif gc < 40:
            print("   Stability  : 🔵 Low GC — less stable")
        else:
            print("   Stability  : 🟢 Normal range")

        print("   RNA        :", rna[:35], "...")
        print("   Rev Comp   :", rev_comp[:35], "...")
        print("-" * 55)


# Run the analysis
analyze_fasta(sequences)


# ---- STEP 5: SAVE RESULTS TO FILE ----

def save_results(sequences, output_file):

    with open(output_file, "w") as out:

        out.write("FASTA ANALYSIS REPORT\n")
        out.write("=" * 50 + "\n")

        for name, dna in sequences.items():

            dna = dna.upper().strip()

            if len(dna) == 0:
                continue

            gc = round(
                (dna.count("G") + dna.count("C")) / len(dna) * 100, 2
            )

            out.write("\nSequence : " + name + "\n")
            out.write("Length   : " + str(len(dna)) + " bases\n")
            out.write("GC       : " + str(gc) + "%\n")
            out.write("-" * 50 + "\n")

    print()
    print("✅ STEP 5: Results saved to", output_file)


save_results(sequences, "analysis_results.txt")


# ---- STEP 6: VERIFY OUTPUT FILE ----

print()
print("✅ STEP 6: Contents of analysis_results.txt:")
print("-" * 40)

with open("analysis_results.txt", "r") as file:
    print(file.read())
```
