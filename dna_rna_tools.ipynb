```python
# COMPLETE DNA ANALYSIS TOOL
# Combines Day 1 + Day 2 + Day 3

def complete_dna_analysis(dna):

    # Clean input
    dna = dna.upper().replace(" ", "").replace("\n", "")

    # Complement dictionary
    complement_map = {
        "A" : "T",
        "T" : "A",
        "G" : "C",
        "C" : "G"
    }

    # --- Day 1 — Nucleotide counts ---
    a = dna.count("A")
    t = dna.count("T")
    g = dna.count("G")
    c = dna.count("C")

    # --- Day 2 — GC content ---
    gc = round((g + c) / len(dna) * 100, 2)

    # --- Day 3 — Transcription ---
    rna = dna.replace("T", "U")

    # --- Day 3 — Complement ---
    complement = ""
    for base in dna:
        complement = complement + complement_map[base]

    # --- Day 3 — Reverse complement ---
    rev_complement = complement[::-1]

    # --- Print full report ---
    print("=" * 50)
    print("      COMPLETE DNA ANALYSIS REPORT")
    print("=" * 50)
    print("DNA Sequence      :", dna)
    print("Length            :", len(dna), "bases")
    print()
    print("-- NUCLEOTIDE COUNTS --")
    print("A:", a, " T:", t, " G:", g, " C:", c)
    print()
    print("-- GC CONTENT --")
    print("GC Content        :", gc, "%")
    if gc > 60:
        print("Stability         : High GC - very stable 🔴")
    elif gc < 40:
        print("Stability         : Low GC - less stable 🔵")
    else:
        print("Stability         : Normal GC range 🟢")
    print()
    print("-- TRANSCRIPTION AND COMPLEMENT --")
    print("RNA               :", rna)
    print("Complement        :", complement)
    print("Reverse Complement:", rev_complement)
    print("=" * 50)

# Run it with a test sequence
complete_dna_analysis("ATGAAAGCAATTTTCGTACTGAAAGGTTTTGTT")
```
